# Example: Titanic Dataset

|Variable|Definition|
|-|-|
|Survived|Wherther a Passenger survive|
|Pclass|Ticket class|
|Sex|Passenger's gender|
|Age|Passenger's age|
|Sibsp|Number of siblings/spouses aboard ship|
|Parch|Number of parents/children aboard ship|
|Fare|Passenger's fare|
|Cabin|Cabin number|
|Embarked|Port of Embarkation|

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
# barplot
plt.figure(figsize=(12,5))

plt.subplot(121)
sns.barplot(data=titanic,x='embark_town',y="survived")

plt.subplot(122)
sns.barplot(data=titanic,x='embark_town',y='survived',hue='sex',ci=None)
```

```python
# countplot
plt.figure(figsize=(12,5))

plt.subplot(121)
sns.countplot(data=titanic,x="survived")

plt.subplot(122)
sns.countplot(data=titanic,x="survived",hue="class")
```

```python
# boxplot
plt.figure(figsize=(12,6))

plt.subplot(121)
sns.boxplot(data=titanic,x="sex",y="age")

plt.subplot(122)
sns.boxplot(data=titanic,x="sex",y="age",hue="survived")
```

```python
# stripplot
plt.figure(figsize=(13,13))

plt.subplot(221)
sns.stripplot(data=titanic,x="sex",y="age")

plt.subplot(222)
sns.stripplot(data=titanic,x="sex",y="age",hue="sibsp")

# SwarmPlot
plt.subplot(223)
sns.swarmplot(data=titanic,x="sex",y="age")

plt.subplot(224)
sns.swarmplot(data=titanic,x="sex",y="age",hue="sibsp")
```

```python
# FacetGrid
g = sns.FacetGrid(titanic, col='survived', row='sex', height=4)
g.map(sns.histplot, 'embarked')
```

```python
# FacetGrid-2
grid = sns.FacetGrid(titanic, col='pclass', row='survived', height=4)
grid.map(sns.kdeplot, 'age', shade=True)
plt.show()
```

# Example: Boston Housing Dataset


|Variable|Definition|
|-|-|
|CRIM|Per capita crime rate by town|
|ZN|Proportion of residential land zoned for lots over 25000 sq. ft.|
|INDUS|Proportion of non-retail business acres per town|
|CHAS|Charles river dummy variable|
|NOX|Nitric oxides concentration|
|RM|Average number of rooms per dwelling|
|AGE|Proportion of owner-occupied units built prior to 1940|
|DIS|Weighted distance to five Boston employment centres|
|RAD|Index of accessibility of radial highway|
|TAX|Full-value property-tax rate|
|PTRATIO|Pupil-teacher ratio by town|
|B|1000(Bk-0.63)^2,Bk is the proportion of blacks by town|
|LSTAT|Lower status of population|
|MEDV|Median value of owner-occupied homes|


```python
from sklearn.datasets import load_boston
boston_data = load_boston()
```

```python
boston = pd.DataFrame(boston_data.data,columns=boston_data.feature_names)
boston["MEDV"] = boston_data.target
boston.head()
```

```python
# scatterplot
plt.figure(figsize=(12,12))

plt.subplot(221)
sns.scatterplot(data=boston,x="LSTAT",y="MEDV")

plt.subplot(222)
sns.scatterplot(data=boston,x="LSTAT",y="MEDV",hue="CHAS")

plt.subplot(223)
sns.scatterplot(data=boston,x="LSTAT",y="MEDV",hue="CHAS",style="RAD")

plt.subplot(224)
sns.scatterplot(data=boston,x="LSTAT",y="MEDV",hue="CHAS",size="RM")
```

```python
# regplot
plt.figure(figsize=(12,6))
sns.regplot(data=boston,x="LSTAT",y="MEDV")
```

```python
# residplot
plt.figure(figsize=(12,6))
sns.residplot(data=boston,x="LSTAT",y="MEDV")
```

```python
# regplot-2
plt.figure(figsize=(12,6))
sns.regplot(data=boston,x="LSTAT",y="MEDV",order=2)
```

```python
# regplot
plt.figure(figsize=(12,5))
sns.jointplot(data=boston,x="RM",y="MEDV",kind="reg")
```

```python
# displot
sns.displot(data=boston,x="MEDV",kde=True)
```

```python
# heatmap
plt.figure(figsize=(10,8))
sns.heatmap(boston.corr(),annot=True)
```

# Example:Iris Dataset

```python
iris = sns.load_dataset('iris')
```

```python
# histplot
plt.figure(figsize=(18,10))

plt.subplot(231)
sns.histplot(data=iris,x="sepal_length")

plt.subplot(232)
sns.histplot(data=iris,x="sepal_length",binwidth=0.8)

plt.subplot(233)
sns.histplot(data=iris,x="sepal_length",kde=True)

plt.subplot(234)
sns.histplot(data=iris,x="sepal_length",hue="species")

plt.subplot(235)
sns.histplot(data=iris,x="sepal_length",hue="species",multiple="stack")

plt.subplot(236)
sns.histplot(data=iris,x="sepal_length",hue="species",multiple="dodge")
```

```python
# kdeplot
plt.figure(figsize=(12,8))

plt.subplot(221)
sns.kdeplot(data=iris,x="sepal_length")

plt.subplot(222)
sns.kdeplot(data=iris,x="sepal_length",cumulative=True)

plt.subplot(223)
sns.kdeplot(data=iris,x="sepal_length",hue="species")

plt.subplot(224)
sns.kdeplot(data=iris,x="sepal_length",hue="species",multiple="stack")
```

```python
# jointplot
sns.jointplot(data=iris,x="sepal_length",y="petal_length",hue="species")
```

```python
# jointplot-2
sns.jointplot(data=iris,x="sepal_length",y="petal_length",hue="species",kind="kde")
```

```python
# Pairplot
sns.pairplot(data=iris,kind="scatter")
```

```python
# Pairplot-2
sns.pairplot(data=iris,kind="scatter",hue="species")
```

```python
# pairgrid
plt.figure(figsize=(10,10))
g = sns.PairGrid(data=iris)
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
```

```python
# pairgrid-2
plt.figure(figsize=(10,10))
g = sns.PairGrid(data=iris,hue="species")
g.map_upper(sns.scatterplot)
g.map_diag(sns.histplot)
g.map_lower(sns.kdeplot)
g.add_legend()
```

```python
# jointgrid
g = sns.JointGrid(data=iris, x="sepal_length", y="sepal_width", hue="species")
g.plot(sns.scatterplot, sns.histplot)
```

```python
# jointgrid-2
g = sns.JointGrid(data=iris,x="sepal_length",y="sepal_width",hue="species")
g.plot(sns.kdeplot,sns.boxplot)
```

```python
# jointgrid-3
g = sns.JointGrid()
x,y = iris["sepal_length"],iris["sepal_width"]
sns.scatterplot(x=x,y=y,ax=g.ax_joint)
sns.boxplot(x=x,ax=g.ax_marg_x)
sns.histplot(y=y,ax=g.ax_marg_y)
```

# Other plots

```python
plt.figure(figsize=(12,12))

plt.subplot(221)
sns.ecdfplot(data=iris,x="sepal_length")

plt.subplot(222)
sns.violinplot(data=iris,x="species",y="sepal_length")

plt.subplot(223)
sns.boxenplot(data=iris)

plt.subplot(224)
sns.pointplot(data=iris,x="species",y="sepal_length")
```

```python
# catplot
plt.figure(figsize=(10,10))
sns.catplot(data=titanic,x="age",y="class")
```

```python
# violin plot
# 類似boxplot,中間的黑線即為boxplot
plt.figure(figsize=(10,10))
sns.violinplot(data=iris,x="species",y="sepal_length")
```

```python
# boxenplot
# 用於更大型的資料,藉由繪製更多個百分位數,可以更清楚知道觀察值的分布
plt.figure(figsize=(10,10))
sns.boxenplot(data=iris)
```

```python
# pointplot
# 顯示點估計與區間估計的結果,實心點為估計的平均,鉛直線代表信賴區間
plt.figure(figsize=(10,10))
sns.pointplot(data=iris,x="species",y="sepal_length")
```

```python
# cluster plot
sns.clustermap(iris.iloc[:,0:4])
```

# Figure-level function vs. Axes-level function

|Figure-level function|Axes-level function|
|-|-|
|FacetGrid物件|Matplotlib-Pyplot-Axes物件|
|圖例在圖形外|圖例在圖形內|
|快速繪製多個子圖|需要手刻所有子圖|
|無法成為更大型圖表的子圖|可以成為更大型圖表的子圖|
|使用kind參數可以畫出想要的圖形|直接用xxxplot畫出想要的圖形|

```python
# relplot
sns.relplot(data=iris,x="sepal_length",y="sepal_width",hue="species",kind="scatter")
```

```python
# relplot
sns.scatterplot(data=iris,x="sepal_length",y="sepal_width",hue="species")
```
